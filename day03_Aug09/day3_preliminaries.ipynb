First we need to import all of the packages we need. In addition to the usual numpy, pandas, and matplotlib packages we will need to use some special functions that are implement in the _scipy_ python package. We will be using the Error Function and Inverse Error Function (https://docs.scipy.org/doc/scipy/reference/generated/scipy.special.erf.html)

```python
# All the imports we need
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf, erfinv
import pandas as pd
```

__Part 1__

In the assignment today you will need to use the inverse error function. We'll start by plotting the error function and inverse error function to see what they look like.

```python
# First define an array of x values

x = np.arange(-10, 10, 0.01) # Array starting at -10, going to 10 in steps of 0.01

# get y arrays where y1 = erf(x) and y2 = erfinv(x)
y1 = erf(x)
y2 = erfinv(x) # NOTE: because of the way we've imported these error functions, we don't need any scipy. before the function
```

```python
# Plot the two error functions along x

# setting up the plot
plt.xlabel("x")
plt.ylabel("y")
plt.grid() #let's have a nice grid on our plot

#plot both functions
plt.plot(x, y1, label="y=erf(x)") #set up labels for y1 and y2
plt.plot(x, y2, label="y=erfinv(x)")
plt.legend() # get a legend to display the labels

plt.show()
```

__Part 2__

In this part we're going to get practice plotting an infinite sum. As an example, we'll use the Taylor series approximation for $e^x$. 
$$e^x = \sum_{n=0}^\infty \frac{x^n}{n!}$$

On a computer we can't truly represent an infinite sum, this would be like having a for loop that goes on forever. But, if we choose a large enough amount of iterations, the value we are summing will be a good approximation for the infinite sum:
$$e^x \approx \sum_{n=0}^N \frac{x^n}{n!}$$
As long as $N>>0$. Try changing $N$ in the code below and see how big $N$ has to be to make the sum a good approximation of $e^x$. 

```python
# As we did before, define an x array which will be our function's domain

x = np.arange(0, 10, 0.1) # an array that goes from 0 to 10 in steps of 0.1

# Now let's define a maximu value for our sum
N = 1 # The maximum value of the sum. PLAY AROUND WITH THIS! What's the smallest N can be to make a good approximation


y_sum = np.zeros(len(x)) # This array will be where we store our sum at the current step. It should be the same length as x, as it will be the value of y at all domain points.

# We can implement the sum as a for loop going up to N
for n in range(N+1): #remember the range function goes from 0 to N-1 so if we want to go to N we need to put in N+1
    #the expression in the sum
    y_sum = y_sum + x**n/np.math.factorial(n)
            
# For comparison, let's make another variable that holds the actual exponential function
y_exp = np.exp(x)

# Now lets plot the two functions to compare

#set up the plot
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

# plot both functions
plt.plot(x, y_exp, label="Numpy's representation of e^x")
plt.plot(x, y_sum, label="Our sum representation of e^x")
plt.legend()
plt.show()
```

__Part 3__ 

In today's assignment you may find you need to make a contour plot. Here's an example of a contour plot, plotting the contours of the function
$$z = \sqrt{x^2 + y^2}$$

```python
# First, make two arrays for x and y

x = np.arange(-10, 10, 0.1) #Both arrays go from -10 to 10 in steps of 0.1
y = np.arange(-10, 10, 0.1)

# Now we need to make something called a mesh grid. This will give us 2D representations of our x and y arrays. For more, see https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html

xx, yy = np.meshgrid(x, y)

# define a variable z with the mesh grid variables
z = np.sqrt(xx**2 + yy**2)

# Make the contour plot

# the set up is the same as before
plt.xlabel("x")
plt.ylabel("y")

# here instead of making the plot with plt.plot(), we make it with plt.contour(), which has similar syntax
cont = plt.contour(xx, yy, z)
plt.clabel(cont, inline=True, fontsize=10) #This line puts the values of each contour onto the plot
plt.show()
```
